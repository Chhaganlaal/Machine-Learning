```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-pastel')
```

```python
data_y = pd.read_csv('E:\Machine Learning\ML_course\Datasets\KNN\datay.csv')
data_x = pd.read_csv('E:\Machine Learning\ML_course\Datasets\KNN\datax.csv')
```

```python
X = data_x.iloc[:,1:].values
Y = data_y.iloc[:,1:].values
Y = Y.reshape((-1,))
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1], c=Y)
plt.show()
```

```python
query = np.array([1,3.5])
plt.scatter(X[:,0],X[:,1], c=Y)
plt.scatter(query[0],query[1], color='r')
plt.show()
```

```python
def distance(x1, x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(x, y, query, k=15):
    vals = []
    m = x.shape[0]
    for i in range(m):
        d = distance(query, x[i])
        vals.append((d, y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    unique_vals = np.unique(vals[:,1], return_counts=True)
    index = unique_vals[1].argmax()
    pred = unique_vals[0][index]
    return pred
    
```

```python
k = int(Y.shape[0]*0.05)
pred = knn(X, Y, query, k)
print("Predicted answer: %s" %pred)
```
